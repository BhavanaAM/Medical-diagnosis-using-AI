```python
heart_data.drop_duplicates(inplace=True)
heart_data.duplicated().sum()
```

```python
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

Accuracy of trained dataset

```python
x_train_prediction=model.predict(x_train)
training_data_r2=r2_score(y_train,x_train_prediction)
training_data_r2
```

```python
training_data_mse = mean_squared_error(y_train, x_train_prediction)
training_data_mse
```

```python
training_data_mae = mean_absolute_error(y_train, x_train_prediction)
training_data_mse
```

```python
input_data=(41,0,1,130,204,0,0,172,0,1.4,2,0,2)
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
print(prediction)
if(prediction[0]==0):
  print('The person does not have heart disease')
else:
  print('The person has heart disease')
```
